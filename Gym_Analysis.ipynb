```python
# Loading dataset
import pandas as pd

data = pd.read_csv('gym_members_exercise_tracking.csv')

# Displaying first few rows
print(data.head())
```

```python
# Getting overview
print(data.info())
```

```python
# Checking for missing values
print(data.isnull().sum())
```

```python
# Summary stats
print(data.describe())
```

```python
# Analyze popular workout types
workout_counts = data['Workout_Type'].value_counts()
print(workout_counts)
```

## Exploring the Data


```python
# Analyze workout frequency by experience level
import matplotlib.pyplot as plt

workout_by_experience = data.groupby('Experience_Level')['Workout_Frequency (days/week)'].mean()
workout_by_experience.plot(kind='bar')
plt.title('Average Workout Frequency by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Workout Frequency (days/week)')
plt.show()
```

```python
import seaborn as sns

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Experience_Level', y='Session_Duration (hours)', data=data, color='b')

# Add labels and title
plt.title('Experience Level vs Session Duration', fontsize=16)
plt.xlabel('Experience Level', fontsize=14)
plt.ylabel('Session Duration (hours)', fontsize=14)

# Show plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Workout_Frequency (days/week)', y='Session_Duration (hours)', data=data, color='r')
plt.title('Session Duration vs Workout Frequency', fontsize=16)
plt.xlabel('Workout Frequency (days/week)', fontsize=14)
plt.ylabel('Session Duration (hours)', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Fat_Percentage', y='Session_Duration (hours)', data=data, color='b')
plt.title('Session Duration vs Fat Percentage', fontsize=16)
plt.xlabel('Fat Percentage', fontsize=14)
plt.ylabel('Session Duration (hours)', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BMI', y='Session_Duration (hours)', data=data, color='m')
plt.title('Session Duration vs BMI', fontsize=16)
plt.xlabel('BMI', fontsize=14)
plt.ylabel('Session Duration (hours)', fontsize=14)
plt.show()
```

# Visualization Creation

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Workout_Type', hue='Gender', data=data)
plt.title('Gender Distribution of Workout Types', fontsize=16)
plt.xlabel('Workout Type', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Workout_Type', y='Avg_BPM', data=data)
plt.title('Workout Type vs Average BPM', fontsize=16)
plt.xlabel('Workout Type', fontsize=14)
plt.ylabel('Average BPM', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='BMI', data=data)
plt.title('BMI Distribution Across Gender', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('BMI', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = data[['Age', 'Fat_Percentage', 'Resting_BPM', 'Workout_Frequency (days/week)', 'Experience_Level', 'BMI', 'Session_Duration (hours)', 'Calories_Burned']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Key Features', fontsize=16)
plt.show()
```

```python
# Group the data by Workout_Type and count the occurrences
activity_counts = data['Workout_Type'].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=activity_counts.index, y=activity_counts.values, palette='viridis')

# Add labels and title
plt.xlabel('Workout Type', fontsize=12)
plt.ylabel('Total Times Done', fontsize=12)
plt.title('Total Times Each Workout Type Was Done', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by Experience_Level and count occurrences
experience_counts = data['Experience_Level'].value_counts()

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=experience_counts.index, y=experience_counts.values, palette='coolwarm')

# Add labels and title
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Total Counts', fontsize=12)
plt.title('Total Counts of Each Experience Level', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter the data for Experience Level 2
experience_level_2 = data[data['Experience_Level'] == 2]

# Group by Workout_Type and count occurrences
activity_counts = experience_level_2['Workout_Type'].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=activity_counts.index, y=activity_counts.values, palette='viridis')

# Add labels and title
plt.xlabel('Workout Type', fontsize=12)
plt.ylabel('Total Counts', fontsize=12)
plt.title('Total Counts of Each Activity for Experience Level 2', fontsize=14)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Filter the data for Experience Level 3
experience_level_3 = data[data['Experience_Level'] == 3]

# Group by Workout_Type and count occurrences
activity_counts = experience_level_3['Workout_Type'].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=activity_counts.index, y=activity_counts.values, palette='coolwarm')

# Add labels and title
plt.xlabel('Workout Type', fontsize=12)
plt.ylabel('Total Counts', fontsize=12)
plt.title('Total Counts of Each Activity for Experience Level 3', fontsize=14)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Group the data by Workout_Type and Experience_Level, and count occurrences
activity_experience_counts = data.groupby(['Workout_Type', 'Experience_Level']).size().reset_index(name='Counts')

# Create the grouped bar plot
plt.figure(figsize=(14, 8))
sns.barplot(data=activity_experience_counts, x='Workout_Type', y='Counts', hue='Experience_Level', palette='viridis')

# Add labels and title
plt.xlabel('Workout Type', fontsize=12)
plt.ylabel('Total Counts', fontsize=12)
plt.title('Total Counts of Each Activity by Experience Level', fontsize=14)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Set up the figure size and style
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Create a histogram for Session_Duration (hours) colored by Experience_Level
sns.histplot(data=data, x='Session_Duration (hours)', hue='Experience_Level', kde=True, bins=20, palette='viridis', element='step')

# Add labels and title
plt.xlabel('Session Duration (hours)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Workout Session Durations by Experience Level', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Grouping the data by Gender and Experience_Level and counting occurrences
gender_experience_counts = data.groupby(['Gender', 'Experience_Level']).size().reset_index(name='Count')

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=gender_experience_counts, 
    x='Experience_Level', 
    y='Count', 
    hue='Gender', 
    palette='viridis'
)

# Adding titles and labels
plt.title('Gender Distribution Across Experience Levels', fontsize=16)
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each experience level
experience_counts = data['Experience_Level'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    experience_counts, 
    labels=experience_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['#FF9999', '#66B3FF', '#99FF99']
)
plt.title('Experience Level Distribution')
plt.show()
```
